````python
import requests
from IPython.display import Audio
import wave

def get_download_url(gdrive_url):
    file_id = gdrive_url.split("/file/d/")[1].split("/")[0]
    download_url = f"https://drive.google.com/uc?id={file_id}"

    return download_url

def play_wav_from_url(url):
    response = requests.get(url)
    with open("audio.wav", "wb") as f:
        f.write(response.content)

    with wave.open("audio.wav", 'rb') as wav_file:
        params = wav_file.getparams()
        audio = wav_file.readframes(params.nframes)
        audio_length = params.nframes / float(params.framerate)
        print(f"length: {audio_length} sec")
        return Audio(audio, rate=params.framerate)

# Google Drive URL
gdrive_url = "https://drive.google.com/file/d/1Ebvw10fItoKEJt2h_Han1g_YkRtE85Jr/preview"
download_url = get_download_url(gdrive_url)
play_wav_from_url(download_url)
````
